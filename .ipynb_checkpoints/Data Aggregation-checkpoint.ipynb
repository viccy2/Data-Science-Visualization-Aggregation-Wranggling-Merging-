```python
import pandas as pd
```

```python
import numpy as np
```

```python
data  = pd.DataFrame(np.random.randn(20, 4), index = pd.date_range('1/1/2021', periods=20), columns = ['A', 'B','C','D'])
```

```python
data
```

```python
r = data.rolling(window=3, min_periods=1)
```

```python
r.aggregate(np.sum)
```

```python
r['A'].aggregate(np.sum)
```

```python
r[['A','B']].aggregate(np.sum)
```
