```python
#Visualization of Binomial Distribution (Discrete Distribution e.g toss of a coin)
```

```python
sns.distplot(random.binomial(n=10, p =0.5, size=1000),hist=True, kde=False)
```

```python
#Poisson Distributio (How many times an event can occur)
```

```python
sns.distplot(random.poisson(lam=2, size=1000), kde=False)
```

```python
#Uniform Distribution (Where every event has probability of occuring)
```

```python
sns.distplot(random.uniform(size=1000), hist=False)
```

```python
#Logistic Distribution (used to describe Growth)
```

```python
sns.distplot(random.logistic(loc=1, scale=2, size=(2,3)))
```

```python
#Zipf Distribution (based on zipf's Law of nth terms)
```

```python
x = random.zipf(a=2, size=1000)
sns.distplot(x[x<10], kde=False)
```
